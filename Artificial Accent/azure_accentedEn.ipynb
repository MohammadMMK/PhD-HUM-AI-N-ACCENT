## accented 1a IPA

θ → t
ð → d
ɹ → r
ɪ → i
æ → a
ɝ → e

```python


# Choose voice
synthesis_voice_name = "en-US-AvaMultilingualNeural"
language = "en-US"
# Output file
output_file = "accentedEnglish1.wav"



text = [
    "The three brothers are reading a very interesting book.",
    "The red cat is sitting near the bird.",
    "I think this weather is very good."
]

ipa = [
    "də tri ˈbradɚz ar ˈridin ə ˈveri ˈintɚestin buk",
    "də red kat iz ˈsitin nir də bed",
    "ai tink dis ˈweder iz ˈveri ɡud"
]


azure_ipa(
    text,
    ipa,
    language,
    synthesis_voice_name,
    output_file,
    pause_ms="700ms"
)
```
