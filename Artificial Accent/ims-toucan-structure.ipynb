# manipulating the IPA


## Manipulating the stress location

```python
# manipulate the stress locations
IPA0 = "~il trˈɛno arɾˈiva.~#" # correct stress locations
IPA1 = "~ˈil ˈtrɛno ˈarɾiva.~#" #wrong stress locations
IPA2 = "~il trɛno arɾiva.~#" # no stress signs
IPAs = [IPA0, IPA1, IPA2]
```

```python
tts.set_phonemizer_language("ita") # we do not need because we input directly the phones
tts.set_accent_language("ita") # this will define the language embeddings used 
for i, phonem in enumerate(IPAs):
    wave, sr = tts(phonem, input_is_phones=True)
    import soundfile as sf
    sf.write(f"outputs/IPA{i}.wav", wave, sr)
```
