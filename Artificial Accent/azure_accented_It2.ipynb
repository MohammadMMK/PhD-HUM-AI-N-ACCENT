# accent 1
substitution of a phoneme with existing phoneme

```python

# Choose voice
synthesis_voice_name = "it-IT-IsabellaNeural"
language = "it-IT"
# Output file
output_file = "accentedItalian_Ssupra.wav"

text = [
    "Il sabato mangio sempre la pizza.",
    "Sette soldati bevono il vino.",
    "Un topo salta sopra il tavolo.",
]

ipa_manipulated = [
    "il saňąbato ňąmand ío ňąs…õmpre la ňąpittsa",
    "ňąs…õtte ňąsoldati ňąbevono il ňąvino",
    "un ňąt…Ēpo ňąsalta ňąs…Ēpra il taňąvolo",
]
azure_ipa(
    text,
    ipa_manipulated,
    language,
    synthesis_voice_name,
    output_file,
    pause_ms="700ms"
)
print (f"speech model: {synthesis_voice_name}")
print (f"text: {text}")
print (f"ipa: {ipa_manipulated}")
```
