# accent 1 
###   [r]  -> [ɹ]   (English-like postalveolar approximant)
###   [ʎʎ] -> [lj]  (de-palatalised, de-geminated)
###   [ɲɲ] -> [nj]  (de-palatalised, de-geminated)
###  [ɛ]  -> [e]   (raising / closing of mid-front)
###   [ɔ]  -> [o]   (raising / closing of mid-back)


```python

# Choose voice
synthesis_voice_name = "it-IT-IsabellaMultilingualNeural"
language = "it-IT"
# Output file
output_file = "accentedItalian1.wav"

text = [
    "Mio figlio sogna una bella montagna rossa.",
    "Ogni notte mia moglie sogna un bel cielo rosso.",
    "Il figlio del signore vuole sette belle uova rosse in campagna.",
]

# Standard (educated central / "neutral") Italian IPA
ipa_manipulated = [
    "ˈmio ˈfiljo ˈsonja ˈuna ˈbella monˈtanja ˈɹossa",
    "ˈonji ˈnotte ˈmia ˈmolje ˈsonja un bel ˈtʃelo ˈɹosso",
    "il ˈfiljo del sinˈjore ˈvwole ˈsette ˈbelle ˈwova ˈɹosse in kamˈpanja",
]
azure_ipa(
    text,
    ipa_manipulated,
    language,
    synthesis_voice_name,
    output_file,
    pause_ms="700ms"
)
print (f"speech model: {synthesis_voice_name}")
print (f"text: {text}")
print (f"ipa: {ipa_manipulated}")
```

# accent 2 
###   [r]  -> [ʁ]   (french-like postalveolar approximant)
###   [ʎʎ] -> [lj]  (de-palatalised, de-geminated)
###   [ɲɲ] -> [nj]  (de-palatalised, de-geminated)
###  [ɛ]  -> [e]   (raising / closing of mid-front)
###   [ɔ]  -> [o]   (raising / closing of mid-back)


```python
trill_variants = {
    "english":         "ɹ",   # postalveolar approximant
    "german_french":   "ʁ",   # voiced uvular fricative
    "german_alt":      "ʀ",   # uvular trill
    "romanian":        "r̪",   # dental trill
    "russian":         "ɾ̪ʲ",  # palatalised apical dental flap
    "romanian_flap":   "ɾ",   # plain alveolar flap
}

def apply_trill(ipa_list, glyph):
    return [s.replace("ɹ", glyph) for s in ipa_list]
```

```python

output_file = "accentedItalian2.wav"
# e.g.
ipa_german = apply_trill(ipa_manipulated, trill_variants["german_french"])
azure_ipa(
    text,
    ipa_german,
    language,
    synthesis_voice_name,
    output_file,
    pause_ms="700ms"
)
print ("german_french")
print (f"speech model: {synthesis_voice_name}")
print (f"text: {text}")
print (f"ipa: {ipa_german}")
```
