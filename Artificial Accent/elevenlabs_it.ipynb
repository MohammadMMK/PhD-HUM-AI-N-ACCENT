# Standard Italian giving IPA

```python
# Standard Italian
input_standard = (
    "Mio /ˈfiʎʎo/ /ˈsoɲɲa/ una /ˈbɛlla/ /monˈtaɲɲa/ /ˈrɔssa/. "
    "/ˈoɲɲi/ /ˈnɔtte/ mia /ˈmɔʎʎe/ /ˈsoɲɲa/ un /bɛl/ /ˈtʃɛlo/ /ˈrɔsso/. "
    "Il /ˈfiʎʎo/ del /siɲˈɲore/ /ˈvwɔle/ /ˈsɛtte/ /ˈbɛlle/ /ˈwɔva/ /ˈrɔsse/ in /kamˈpaɲɲa/."
)
voice_id="gfKKsLN1k0oYYN9n2dXX"# Violetta - Ringing, Bright and Joyful Violetta - A warm, clear, and engaging middle-aged Italien female voice. With a smooth, natural tone, this voice is perfect for narrations, audiobooks, or informative audio content. Its expressive quality makes it ideal for storytelling, while remaining professional and easy to follo

import os
from dotenv import load_dotenv
from elevenlabs.client import ElevenLabs
from elevenlabs.play import play
from IPython.display import Audio, display
load_dotenv()
api_key = os.getenv("ELEVENLABS_API_KEY")
client = ElevenLabs(api_key=api_key)
audio_stream = client.text_to_speech.convert(
    text=input_standard,
    voice_id=voice_id,  
    model_id="eleven_v3",
    output_format="mp3_44100_128"
)
# ElevenLabs returns an iterator/generator, so we combine it into a single bytes object
audio_bytes = b"".join(audio_stream)
# 5. Display the interactive audio player block in Jupyter
print(input_standard)
# display(Audio(audio_bytes, autoplay=False))
# Save the file into your repository folder
with open("elevenlabs_standard_it.mp3", "wb") as f:
    f.write(audio_bytes)
```

# Accent 1
### German-accented

| Feature | Standard token | German token |
| --- | --- | --- |
| [r] → [ʁ] | /ˈrɔssa/, /ˈrɔsso/, /ˈrɔsse/ | /ˈʁossa/, /ˈʁosso/, /ˈʁosse/ |
| [ʎʎ] → [lj] | /ˈfiʎʎo/, /ˈmɔʎʎe/ | /ˈfiljo/, /ˈmolje/ |
| [ɲɲ] → [nj] | /ˈsoɲɲa/, /monˈtaɲɲa/, /ˈoɲɲi/, /siɲˈɲore/, /kamˈpaɲɲa/ | /ˈsonja/, /monˈtanja/, /ˈonji/, /sinˈjore/, /kamˈpanja/ |
| [ɛ] → [e] | /ˈbɛlla/, /bɛl/, /ˈtʃɛlo/, /ˈsɛtte/, /ˈbɛlle/ | /ˈbella/, /bel/, /ˈtʃelo/, /ˈsette/, /ˈbelle/ |
| [ɔ] → [o] | /ˈrɔssa/, /ˈnɔtte/, /ˈmɔʎʎe/, /ˈrɔsso/, /ˈvwɔle/, /ˈwɔva/, /ˈrɔsse/ | /ˈʁossa/, /ˈnotte/, /ˈmolje/, /ˈʁosso/, /ˈvwole/, /ˈwova/, /ˈʁosse/ |
