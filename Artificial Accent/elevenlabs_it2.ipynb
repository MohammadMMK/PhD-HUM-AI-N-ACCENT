# Accent 1
### German-accented

| Feature | Standard token | German token |
| --- | --- | --- |
| [r] → [ʁ] | /ˈrɔssa/, /ˈrɔsso/, /ˈrɔsse/ | /ˈʁossa/, /ˈʁosso/, /ˈʁosse/ |
| [ʎʎ] → [lj] | /ˈfiʎʎo/, /ˈmɔʎʎe/ | /ˈfiljo/, /ˈmolje/ |
| [ɲɲ] → [nj] | /ˈsoɲɲa/, /monˈtaɲɲa/, /ˈoɲɲi/, /siɲˈɲore/, /kamˈpaɲɲa/ | /ˈsonja/, /monˈtanja/, /ˈonji/, /sinˈjore/, /kamˈpanja/ |
| [ɛ] → [e] | /ˈbɛlla/, /bɛl/, /ˈtʃɛlo/, /ˈsɛtte/, /ˈbɛlle/ | /ˈbella/, /bel/, /ˈtʃelo/, /ˈsette/, /ˈbelle/ |
| [ɔ] → [o] | /ˈrɔssa/, /ˈnɔtte/, /ˈmɔʎʎe/, /ˈrɔsso/, /ˈvwɔle/, /ˈwɔva/, /ˈrɔsse/ | /ˈʁossa/, /ˈnotte/, /ˈmolje/, /ˈʁosso/, /ˈvwole/, /ˈwova/, /ˈʁosse/ |
